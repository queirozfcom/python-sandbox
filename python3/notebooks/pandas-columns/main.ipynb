## apply vs map

```python
df = pd.DataFrame({
    'name': ['alice','bob','charlie'],
    'age': [25,26,27],
    'number_of_children':[1,3,4]
})[['name','age','number_of_children']]
df
```

```python
df[['name']].apply(lambda row: row.str.capitalize())
```

```python
df[['age','number_of_children']].apply(lambda row: row *2)
```
