## truncation

```python
np.array(range(1500))
```

```python
np.set_printoptions(threshold=5000)
np.array(range(1500))
```

## Show more edge items

```python
# reset to default first
np.set_printoptions(threshold=1000)
```

```python
np.array(range(1500))
```

```python
np.set_printoptions(edgeitems=5)
np.array(range(1500))
```

## setting options temporarily

```python
import numpy as np

with np.printoptions(edgeitems=1):
    # numpy code here uses the given options
    print(np.array(range(1500)))

# other code is under current options
print(np.array(range(1500)))
```
